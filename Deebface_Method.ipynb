### Real time video demo for Face Emotion Recognition

```python
# Kamera-Test
import cv2
import matplotlib.pyplot as plt

# Laden der Kamera
cap = cv2.VideoCapture(0)

# berprfen, ob die Kamera erfolgreich geffnet wurde
if not cap.isOpened():
    raise IOError("Kann die Kamera nicht ffnen")

# Erfassen eines Einzelbildes
ret, frame = cap.read()

# berprfen, ob das Einzelbild erfolgreich erfasst wurde
if not ret:
    raise IOError("Fehler beim Erfassen des Bildes von der Kamera")

# Speichern des Bildes als Datei
cv2.imwrite("foto.jpg", frame)

# Anzeigen des Bildes mit Matplotlib
img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis('off')  # Deaktivieren der Achsenbeschriftungen
plt.show()

# Freigeben der Kamera
cap.release()
```

```python
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace

font = cv2.FONT_HERSHEY_SIMPLEX

# Kamera ffnen
cap = cv2.VideoCapture(0)

# berprfen, ob die Kamera erfolgreich geffnet wurde
if not cap.isOpened():
    raise IOError("Kann die Kamera nicht ffnen")

# Erfassen eines Einzelbildes
ret, frame = cap.read()

# berprfen, ob das Einzelbild erfolgreich erfasst wurde
if not ret:
    raise IOError("Fehler beim Erfassen des Bildes von der Kamera")

# Freigeben der Kamera
cap.release()

# Laden des Haar-Cascade-Klassifikators
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Konvertieren in Graustufen
gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

# Gesichtserkennung
faces = faceCascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=4)

# Emotionserkennung
predictions = DeepFace.analyze(frame)

# Extrahieren der dominanten Emotion aus der Liste 'predictions'
dominant_emotion = predictions[0]['dominant_emotion']

# Zeichnen des Rechtecks um die Gesichter und Anzeigen der Emotion
for (x, y, w, h) in faces:
    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
    cv2.putText(frame,
                dominant_emotion,
                (x, y - 10),  # Position des Textes ber dem Rechteck
                font, 1,
                (0, 255, 0),
                2,
                cv2.LINE_4)

# Anzeigen des Bildes mit Rechteck und Emotion
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
